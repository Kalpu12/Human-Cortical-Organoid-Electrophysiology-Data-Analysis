## Calculate the average velocity over time
# Velocity: slope - y2-y1/x2-x1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv('Data 1.csv')  # Replace with your file path
time = df['Time (hr)'].values
values = df['Condition#1 VentralOnly Green'].values

# Compute cumulative average velocity more efficiently
time_intervals = time[1:] - time[0]  # Calculate all time intervals at once
value_differences = values[1:] - values[0]  # Calculate all value differences at once
avg_velocity_cumulative = value_differences / time_intervals  # Vectorized division

# Create the plot
plt.figure(figsize=(10, 6))

# Plot with improved formatting
plt.plot(time[1:], avg_velocity_cumulative,
         color='gold', linestyle='--', linewidth=2,
         marker='o', markersize=6, label='Cummulative Avg Velocity')

# Add labels and title with improved formatting
plt.xlabel("Time (hr)", fontsize=12)
plt.ylabel("Velocity (movement/hr)", fontsize=12)
plt.title("Cummulative Average Velocity\nCondition #1 Ventral Only Green",
          fontsize=14, pad=20)

# Set axis limits and grid
plt.grid(True)

# Add legend and adjust layout
plt.legend(fontsize=10, framealpha=1)
plt.tight_layout()

# Show plot
plt.show()
```

# Box Plot: Comparison of Velocity for each condition


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle  # Import Rectangle if not already done

# Load the Excel file
df = pd.read_excel('Data 1.xlsx', sheet_name='Data 1')

# Clean column names by removing extra spaces
df.columns = [col.strip() for col in df.columns]

# Define the conditions dictionary
conditions = {
    # 1. Ventral only (V)
    'Ventral None (V)': [
        'Velocity C1: Ventral Only',
        'Velocity C1: VentralOnly',
        'Velocity C1: Ventral Only',
        'Velocity: C1 Ventral Only'
    ],
    # 2. Ventral -> Dorsal (dVd)
    'Ventral -> Dorsal (dVd)': [
        'Velocity: C2 VentralCenter Green',
        'Velocity: C2 Ventral Center Green'
    ],
    # 3. Ventral -> Dorsal (vDv)
    'Ventral -> Dorsal (vDv)': [
        'Velocity: C4 Dorsal Center Red'
    ],
    # 4. Ventral -> Media (dVd)
    'Ventral -> Media (dVd)': [
        'Velocity: C2 VentralCenter towards Media'
    ],
    # 5. Dorsal Only (D)
    'Dorsal None (D)': [
        'Velocity: C3 Dorsal Only Red'
    ],
    # 6. Dorsal -> Ventral (dVd)
    'Dorsal -> Ventral (dVd)': [
        'C2 VentralCenter Red'
    ],
    # 7. Dorsal -> Ventral (vDv)
    'Dorsal -> Ventral (vDv)': [
        'Velocity C4:DorsalCenter Green'
    ],
    # 8. Dorsal -> Media (vDv)
    'Dorsal -> Media (vDv)': [
        'Velocity: C4 DorsalCenter towards Media'
    ]
}

# Function to remove outliers using IQR method
def remove_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    return data[(data >= lower_bound) & (data <= upper_bound)]

# Prepare data for box plots
plot_data = []
clean_condition_names = []

for condition, velocity_cols in conditions.items():
    # Get valid columns that exist in the dataframe
    valid_cols = [col for col in velocity_cols if col in df.columns]

    if not valid_cols:
        continue

    # Combine all replicates for this condition and remove NaN values
    condition_velocities = df[valid_cols].values.flatten()
    condition_velocities = condition_velocities[~np.isnan(condition_velocities)]

    # Remove outliers
    clean_velocities = remove_outliers(condition_velocities)

    if len(clean_velocities) > 0:
        plot_data.append(clean_velocities)
        clean_condition_names.append(condition)

# Create box plot with custom colors
plt.figure(figsize=(14, 8))
ax = plt.gca()

# Custom color scheme
box_colors = ['green', 'red', 'lightgreen', 'orange', 'darkred', 'lime', 'pink', 'lightblue']

# Create box plots
box = ax.boxplot(plot_data, patch_artist=True, showfliers=False)

# Apply custom colors to boxes
for patch, color in zip(box['boxes'], box_colors[:len(plot_data)]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

# Customize whiskers, caps, etc.
for element in ['whiskers', 'caps', 'medians']:
    plt.setp(box[element], color='black', linewidth=1.5)

# Customize the plot
plt.title('Velocity Distribution by Experimental Condition', fontsize=16, pad=20)
plt.ylabel('Velocity', fontsize=14)
plt.xticks(range(1, len(clean_condition_names)+1), clean_condition_names,
           rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add grid and adjust layout
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()

# legend format
# Make sure we have enough colors for all conditions
box_colors = ['green', 'red', 'lightgreen', 'orange', 'darkred', 'lime', 'pink', 'lightblue']

# Create legend elements for ALL conditions (even if some have no data)
legend_elements = [Rectangle((0,0),1,1, fc=box_colors[i % len(box_colors)], alpha=0.7)
                  for i in range(len(conditions))]

plt.legend(legend_elements, conditions.keys(),
           bbox_to_anchor=(1.05, 1),
           loc='upper left',
           fontsize=10,
           title="Conditions")

plt.tight_layout()
plt.show()


```

## Box Plot

this can be used to compare the average number of cells moved across from different cell plates across the different conditions

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# Load data
df = pd.read_excel('Data 1.xlsx')

# Clean column names by removing extra spaces
df.columns = [col.strip() for col in df.columns]

# Define conditions and their corresponding column prefixes
conditions = {
    # 1. Ventral only (V)
    'Ventral None (V)': 'Condition#1 VentralOnly Green',

    # 2. Ventral -> Dorsal (dVd)
    'Ventral -> Dorsal (dVd)': 'Condition#2 VentralCenter Green',

    # 3. Ventral -> Dorsal (vDv)
    'Ventral -> Dorsal (vDv)': 'Condition#4 Dorsal Center Green',

    # 4. Ventral -> Media (dVd)
    'Ventral -> Media (dVd)': 'Condition#2 VentralCenter toward media',

    # 5. Dorsal Only (D)
    'Dorsal None (D)': 'Condition#3 Dorsal Only Red',

    # 6. Dorsal -> Ventral (dVd)
    'Dorsal -> Ventral (dVd)': 'Condition#2 VentralCenter Red',

    # 7. Dorsal -> Ventral (vDv)
    'Dorsal -> Ventral (vDv)': 'Condition#4 Dorsal Center Green',
    #Condition#2 VentralCenter Green

    # 8. Dorsal -> Media (vDv)
    'Dorsal -> Media (vDv)': 'Condition#4 Dorsal Center Red toward Media'
}

# Prepare data - extract all condition columns and remove NaNs
condition_data = {}
for cond_name, prefix in conditions.items():
    # Find all columns that contain the prefix (more flexible than startswith)
    cols = [col for col in df.columns if prefix in col]
    if cols:
        values = df[cols].values.flatten()
        condition_data[cond_name] = values[~np.isnan(values)]

# Filter out conditions with no data
valid_conditions = {k: v for k, v in condition_data.items() if len(v) > 0}

# Set up plot
plt.figure(figsize=(16, 8))
ax = plt.gca()

# Custom color scheme with enough colors for all conditions
box_colors = ['green', 'red', 'lightgreen', 'orange', 'darkred', 'lime', 'pink', 'lightblue']

# Create a box plot for each valid condition
boxes = []
for i, (cond_name, values) in enumerate(valid_conditions.items()):
    if len(values) > 0:
        # Use modulo to ensure we don't exceed color list length
        color_idx = i % len(box_colors)
        box = ax.boxplot(values,
                        positions=[i],
                        patch_artist=True,
                        widths=0.6,
                        showfliers=False,
                        boxprops=dict(facecolor=box_colors[color_idx],
                                    edgecolor='black',
                                    alpha=0.7),
                        whiskerprops=dict(color='black', linewidth=1),
                        capprops=dict(color='black', linewidth=1),
                        medianprops=dict(color='black', linewidth=1.5))
        boxes.append(box)

# Customize plot only if we have valid data
if valid_conditions:
    ax.set_xticks(range(len(valid_conditions)))
    ax.set_xticklabels(valid_conditions.keys(), rotation=45, ha='right', fontsize=12)
    plt.title("Comparison of Cell Movement Across All Conditions", fontsize=16, pad=20)
    plt.ylabel("Axon Displacement", fontsize=14)
    plt.xlabel("Experimental Condition", fontsize=14)
    plt.grid(True, axis='y', linestyle='--', alpha=0.3)

    # Add legend for condition colors
    legend_elements = [Rectangle((0,0),1,1, fc=box_colors[i % len(box_colors)], alpha=0.7)
                     for i in range(len(valid_conditions))]
    plt.legend(legend_elements, valid_conditions.keys(),
              bbox_to_anchor=(1.05, 1),
              loc='upper left',
              fontsize=10,
              title="Conditions")

    plt.tight_layout()
    plt.show()
else:
    print("No valid data found for any condition.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# Load data
df = pd.read_excel('Data 1.xlsx')

# Define time periods for each day
day1 = (0, 24)    # Day 1
day2 = (24, 48)   # Day 2
day3 = (48, 72)   # Day 3
day4 = (72, 96)   # Day 4

# Conditions with colors
conditions = {
    'Ventral None (V)': {
        'prefix': 'Condition#1 VentralOnly Green',
        'color': 'green'
    },
    'Ventral -> Dorsal (dVd)': {
        'prefix': 'Condition#2 VentralCenter Green',
        'color': 'red'
    },
    'Ventral -> Dorsal (vDv)': {
        'prefix': 'Condition#4 Dorsal Center Green',
        'color': 'lightgreen'
    },
    'Ventral -> Media (dVd)': {
        'prefix': 'Condition#2 VentralCenter toward media',
        'color': 'orange'
    },
    'Dorsal None (D)': {
        'prefix': 'Condition#3 Dorsal Only Red',
        'color': 'darkred'
    },
    'Dorsal -> Ventral (dVd)': {
        'prefix': 'Condition#2 VentralCenter Red',
        'color': 'lime'
    },
    'Dorsal -> Ventral (vDv)': {
        'prefix': 'Condition#4 Dorsal Center Red',
        'color': 'pink'
    },
    'Dorsal -> Media (vDv)': {
        'prefix': 'Condition#4 Dorsal Center Red toward Media',
        'color': 'lightblue'
    }
}

plt.figure(figsize=(20, 10))  # Slightly wider to accommodate all days

# Plot boxplots
for i, (cond_name, cond_info) in enumerate(conditions.items()):
    cols = [col for col in df.columns if cond_info['prefix'] in col]
    if not cols:
        continue

    # Day masks (now includes day4)
    masks = [
        (df['Time (hr)'] >= day1[0]) & (df['Time (hr)'] < day1[1]),
        (df['Time (hr)'] >= day2[0]) & (df['Time (hr)'] < day2[1]),
        (df['Time (hr)'] >= day3[0]) & (df['Time (hr)'] < day3[1]),
        (df['Time (hr)'] >= day4[0]) & (df['Time (hr)'] < day4[1])
    ]

    # Plot each day
    for day_idx, mask in enumerate(masks):
        data = df.loc[mask, cols].values.flatten()
        data = data[~np.isnan(data)]

        if len(data) > 0:
            # Updated alpha values for 4 days
            alpha = [1.0, 0.8, 0.6, 0.4][day_idx]  # Different transparency for each day
            plt.boxplot(data,
                       positions=[i*5 + day_idx],  # Changed from 4 to 5 to space out the days
                       widths=0.6,
                       patch_artist=True,
                       boxprops=dict(facecolor=cond_info['color'],
                                   alpha=alpha,
                                   edgecolor='black'),
                       medianprops=dict(color='black', linewidth=2),
                       whiskerprops=dict(color='black'),
                       capprops=dict(color='black'),
                       showfliers=False)

# Customize plot
plt.title('Cell Movement Comparison Across Conditions (4 Days)', fontsize=18, pad=20)
plt.ylabel('Axon Displacement', fontsize=14)

# X-ticks and labels - adjusted for 5 positions per condition
xticks = [i*5 + 1.5 for i in range(len(conditions))]  # Centered in the middle of 4 days
plt.xticks(xticks, conditions.keys(), rotation=0, ha='right', fontsize=10)
plt.yticks(fontsize=12)

# Day indicators - now includes Day 4
y_min, y_max = plt.ylim()
day_label_y = y_min - 0.05 * (y_max - y_min)
for i in range(len(conditions)):
    for day_idx, day_label in enumerate(['Day 1', 'Day 2', 'Day 3', 'Day 4']):
        plt.text(i*5 + day_idx, day_label_y, day_label,  # Changed from 4 to 5
                ha='center', va='top', fontsize=8, color='black')

# Create proper legend using condition colors
legend_elements = [Patch(facecolor=cond_info['color'],
                        label=cond_name,
                        alpha=0.7)  # Use middle alpha for legend
                  for cond_name, cond_info in conditions.items()]

plt.legend(handles=legend_elements,
           bbox_to_anchor=(1.05, 1),
           loc='upper left',
           fontsize=10)

plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch
from itertools import combinations
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

# ============================================================================
# STANDARDIZED PLOT CONFIGURATION
# ============================================================================
PLOT_CONFIG = {
    'figsize': (16, 8),
    'title_fontsize': 18,
    'title_weight': 'bold',
    'title_pad': 20,
    'label_fontsize': 14,
    'tick_fontsize': 12,
    'xtick_fontsize': 10,
    'legend_fontsize': 9,
    'box_width': 0.6,
    'line_width': 1.5,
    'median_width': 2,
    'dpi': 300,
    'alpha': 0.7,
    'grid_alpha': 0.3
}

# Load the Excel file
df = pd.read_excel('Data 1.xlsx', sheet_name='Data 1')

# Clean column names by removing extra spaces
df.columns = [col.strip() for col in df.columns]

# Define time periods for each day (each day = 24 hours)
day_ranges = {
    'day1': (0, 24),
    'day2': (24, 48),
    'day3': (48, 72),
    'day4': (72, 96)
}
day_labels = list(day_ranges.keys())

# Define the conditions dictionary with color information
conditions = {
    'Ventral None (V)': {
        'columns': [
            'Velocity C1: Ventral Only',
            'Velocity C1: VentralOnly',
            'Velocity: C1 Ventral Only'
        ],
        'color': 'green'
    },
    'Ventral -> Dorsal (dVd)': {
        'columns': [
            'Velocity: C2 VentralCenter Green',
            'Velocity: C2 Ventral Center Green'
        ],
        'color': 'red'
    },
    'Ventral -> Dorsal (vDv)': {
        'columns': [
            'Velocity: C4 Dorsal Center Red'
        ],
        'color': 'lightgreen'
    },
    'Ventral -> Media (dVd)': {
        'columns': [
            'Velocity: C2 VentralCenter towards Media'
        ],
        'color': 'orange'
    },
}

# Function to remove outliers using IQR method with absolute values
def remove_outliers(data):
    data = np.asarray(data)
    data = np.abs(data)  # Take absolute value
    if len(data) == 0:
        return np.array([])
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    return data[(data >= lower_bound) & (data <= upper_bound)]

# ======================
# STATISTICAL HELPERS
# ======================
def p_to_stars(p):
    if p < 0.0001:
        return "****"
    elif p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return "ns"

# ======================
# DATA EXTRACTION
# ======================
all_data = {cond: {} for cond in conditions}
sample_sizes = {cond: {} for cond in conditions}

for cond_name, cond_info in conditions.items():
    cols = []
    for col_pattern in cond_info['columns']:
        matching_cols = [col for col in df.columns if col_pattern in col]
        cols.extend(matching_cols)

    if not cols:
        print(f" Warning: No matching columns found for '{cond_name}'.")
        for day_key in day_labels:
            all_data[cond_name][day_key] = np.array([])
            sample_sizes[cond_name][day_key] = 0
        continue

    for day_key, (t0, t1) in day_ranges.items():
        subset = df[(df['Time (hr)'] >= t0) & (df['Time (hr)'] < t1)][cols]
        values = subset.values.flatten()
        values = values[~np.isnan(values)]
        values = remove_outliers(values)
        all_data[cond_name][day_key] = values
        sample_sizes[cond_name][day_key] = len(values)

# ======================
# PLOTTING
# ======================
fig = plt.figure(figsize=PLOT_CONFIG['figsize'])
ax = plt.gca()

conditions_list = list(conditions.keys())
num_conditions = len(conditions_list)

# Define alpha values for each day
day_alphas = [1.0, 0.7, 0.4, 0.2]

# For each condition, extract data for each day and create boxplots
for i, (cond_name, cond_info) in enumerate(conditions.items()):
    for day_index, day_key in enumerate(day_labels):
        data = all_data[cond_name][day_key]

        if len(data) > 0:
            position = i * 5 + day_index  # 5 units spacing between conditions

            plt.boxplot(data, positions=[position], widths=0.6, patch_artist=True,
                         boxprops=dict(facecolor=cond_info['color'], alpha=day_alphas[day_index], edgecolor='black'),
                         medianprops=dict(color='black', linewidth=2),
                         whiskerprops=dict(color='black', linewidth=1.5),
                         capprops=dict(color='black', linewidth=1.5),
                         showfliers=False)

# Customize the plot
plt.title('Ventral Velocity (Absolute) Separate Daily Young Organoids',
          fontsize=18, fontweight='bold', pad=20)
plt.ylabel('Velocity (m/min)', fontsize=14)

# Set x-ticks and labels
xticks = [i * 5 + 1.5 for i in range(num_conditions)]
plt.xticks(xticks, conditions_list, fontsize=10)
plt.yticks(fontsize=12)

# ======================
# STATISTICAL TESTING
# ======================
all_p_values = []
all_comparisons = []

condition_abbrev = {
    'Ventral None (V)': 'V',
    'Ventral -> Dorsal (dVd)': 'Ventral->Dorsal (dVd)',
    'Ventral -> Dorsal (vDv)': 'Ventral->Dorsal (vDv)',
    'Ventral -> Media (dVd)': 'Ventral->Media (dVd)'
}

for day_index, day_key in enumerate(day_labels):
    for condA, condB in combinations(conditions_list, 2):
        dataA = all_data[condA][day_key]
        dataB = all_data[condB][day_key]
        if len(dataA) > 0 and len(dataB) > 0:
            stat, p_value = mannwhitneyu(dataA, dataB, alternative='two-sided')
            all_p_values.append(p_value)
            all_comparisons.append({
                'day_index': day_index + 1,
                'day_key': day_key,
                'condA': condition_abbrev[condA],
                'condB': condition_abbrev[condB],
                'p_value': p_value
            })

# Apply Bonferroni correction
p_value_text_lines = []

if len(all_p_values) > 0:
    reject, p_corrected, _, _ = multipletests(all_p_values, method='bonferroni')

    p_value_text_lines.append("Statistical Comparisons (Bonferroni-corrected):")
    p_value_text_lines.append("-" * 45)

    for day_num in range(1, 5):
        day_comps = [(i, comp) for i, comp in enumerate(all_comparisons) if comp['day_index'] == day_num]

        if day_comps:
            p_value_text_lines.append(f"\nDay {day_num}:")
            for idx, comp in day_comps:
                p_corr = p_corrected[idx]
                stars = p_to_stars(p_corr)

                if p_corr < 0.001:
                    p_str = f"p < 0.001"
                else:
                    p_str = f"p = {p_corr:.3f}"

                comp_label = f"{comp['condA']} vs {comp['condB']}"
                p_value_text_lines.append(f"  {comp_label:30s} {p_str:12s} {stars}")

# ======================
# LEGENDS
# ======================
# Condition legend
legend_elements = [Patch(facecolor=cond_info['color'], label=cond_name)
                  for cond_name, cond_info in conditions.items()]

legend1 = plt.legend(handles=legend_elements,
                    bbox_to_anchor=(1.05, 1),
                    loc='upper left',
                    title='Conditions',
                    fontsize=9,
                    title_fontsize=10)

ax.add_artist(legend1)

# Day legend
day_legend = [Patch(facecolor='gray', alpha=day_alphas[i], label=f'Day {i+1}')
              for i in range(4)]

legend2 = plt.legend(handles=day_legend,
                    bbox_to_anchor=(1.05, 0.75),
                    loc='upper left',
                    title='Time Period',
                    fontsize=9,
                    title_fontsize=10)

ax.add_artist(legend2)

# P-value text box
if p_value_text_lines:
    p_value_text = '\n'.join(p_value_text_lines)

    props = dict(boxstyle='round', facecolor='wheat', alpha=0.15, edgecolor='black', linewidth=0.5)
    ax.text(1.05, 0.48, p_value_text,
            transform=ax.transAxes,
            fontsize=7,
            verticalalignment='top',
            fontfamily='monospace',
            bbox=props)

# Add grid and adjust layout
plt.grid(True, axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.subplots_adjust(right=0.72)

plt.show()

# ======================
# PRINT STATISTICAL SUMMARY
# ======================
print("\n" + "="*80)
print("VENTRAL VELOCITY ANALYSIS COMPLETE")
print("="*80)
print(f"\nTotal comparisons: {len(all_p_values)}")
if len(all_p_values) > 0:
    print(f"Significant after Bonferroni correction: {sum(reject)}")

print("\nSample sizes by condition and day:")
print("-"*80)
for cond_name in conditions_list:
    print(f"\n{cond_name}:")
    for day_key in day_labels:
        n = sample_sizes[cond_name][day_key]
        print(f"  {day_key}: n={n}")

if p_value_text_lines:
    print("\n" + "="*80)
    print("P-VALUE SUMMARY (Bonferroni-corrected):")
    print("="*80)
    print('\n'.join(p_value_text_lines))
    print("="*80)


```
